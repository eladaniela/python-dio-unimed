```python
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Salvador.xlsx")
```

```python
df1.head()
```

```python
#Juntando todos os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes 
```

```python
df.head()
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
#Apagando linhas com valores nulos
df.dropna(inplace = True)
```

```python
#Apagando as linhas com valores nulos de 1 coluna
df.dropna(subset=["Vendas"], inplace = True)
```

```python
#Remover linhas que estejam com valores faltantes em todas as colunas
df.dropna(how ="all", inplace = True)
```

```python
#Criando novas colunas
df["Receitas"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receitas"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receitas"].max()
```

```python
df["Receitas"].min()
```

```python
#Top 3 maiores Receitas
df.nlargest(3, "Receitas")
```

```python
#Top 3 menores Receitas
df.nsmallest(3, "Receitas")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receitas"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receitas", ascending = False).head(10)
```

```python
#Trabalhando com datas
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
#Transformando uma coluna de Data para Data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por Ano
df.groupby(df["Data"].dt.year)["Receitas"].sum()
```

```python
#Criando a nova planilha com ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```
